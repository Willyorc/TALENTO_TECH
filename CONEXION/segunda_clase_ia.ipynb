```python
print('Ejemplo Calculadora Sencilla (+,-,*,/) por WILLIAM CALDERON:')
'''
ESTO ES UN COMENTARIO MULTILINEA
print('Ejemplo Calculadora Sencilla (+,-,*,/) por WILLIAM CALDERON:\n')
PODEMOS DESCOMENTAR Y PROBAR EL print ANTERIOR
'''
import math

print()
#MENU:
print("MENU:")
print("1: Sumar")
print("2: Restar")
print("3: Multiplicar")
print("4: Dividir")
print("5: Potenciación")
print("6: Radicación")
print("7: Logaritmo en base 10")
print("8: Seno")
print("9: Coseno")
print("10: Tangente")

print()

opcion=input("Ingrese la opción deseada del menú: ")

numero_uno=float(input('Ingrese el primer número de la operación:'))
numero_dos=float(input('Ingrese el segundo número de la operación:'))

if opcion == "1":
  print("Vamos a sumar(+) los dos valores ingresados:")
  suma = numero_uno+numero_dos
  print('El resutado es:',suma)

if opcion == '2':
  print("Vamos a restar(-) los dos valores ingresados:")
  resta=numero_uno-numero_dos
  print("El Resultado es:",resta)

if opcion=='3':
  print("Vamos a multiplicar(*) los dos valores ingresados:")
  mult=numero_uno*numero_dos
  print("La multiplicación es: ",mult)

if opcion=='4':
  print("Vamos a dividir(/) los dos valores ingresados:")
  div=numero_uno/numero_dos
  print("La división es: ",div)


if opcion=='5':
  print("Vamos a encontrar Potenciación")
  print("Potenciación:", numero_uno**numero_dos)


if opcion=='6':
  print("Vamos a encontrar Radicación")
  print("Radicación:", math.sqrt(numero_uno))
  print("Radicación:", math.sqrt(numero_dos))


if opcion=='7':
  print("Vamos a encontrar el logaritmo en base")
  print("Logaritmo de numero 1:", math.log10(numero_uno))
  print("Logaritmo de numero 2:", math.log10(numero_dos))

if opcion=='8':
  print("Vamos a encontrar el Seno")
  print("Seno de numero 1: ", math.sin(numero_uno))
  print("Seno de numero 2: ", math.sin(numero_dos))

if opcion=='9':
  print("Vamos a encontrar el Coseno")
  print("Coseno de numero 1: ", math.cos(numero_uno))
  print("Coseno de numero 2: ", math.cos(numero_dos))


if opcion=='10':
  print("Vamos a encontrar la Tangente")
  print("Tangente de numero 1: ", math.tan(numero_uno))
  print("Tangente de numero 2: ", math.tan(numero_dos))
```
